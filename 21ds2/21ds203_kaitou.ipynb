# 最小二乗法

+ 問題　$(x_i,y_i)$ ($0\leq i< N$)が与えられた時、$y=f(x)$で、誤差$\sum_i (f(x_i)-y_i))^2$ が最小となる一次関数$f(x)=ax+b$を求めよ.

# 例題1. $N=2$の場合

## 高校の復習
+  $y=\frac{y_1-y_0}{x_1-x_0}(x-x_0)+y_0=\frac{y_1-y_0}{x_1-x_0}x+\frac{-x_1y_0+x_0y_1}{x_1-x_0}$


## 線形代数を使う

+ $A:=\left[\begin{matrix}x_0&1\\x_1&1\end{matrix}\right]$, $k:=\left[\begin{matrix}a\\b\end{matrix}\right]$, $y:=\left[\begin{matrix}y_0\\y_1\end{matrix}\right]$
とおくと$Ak=y$.

  よって $k=A^{-1}y$

+ 計算すると
  $A^{-1}y=$ $\frac{1}{x_0-x_1}$ $\left[\begin{matrix}1&-1\\-x_1&x_0\end{matrix}\right]$
  $\left[\begin{matrix}y_0\\y_1\end{matrix}\right]$
 $=\frac1{x_0-x_1}\left[\begin{matrix}y_0-y_1\\-x_1y_0+x_0y_1\end{matrix}\right]$


## 最小二乗法を使う

+  ${}^tAAk={}^tAy$ と変形し、$k=({}^tAA)^{-1}{}^tAy$.
+ $N=2$の場合は $A$ は正則なので、
 
 　$k=({}^tAA)^{-1}{}^tAy=A^{-1}({}^tA)^{-1}({}^tA)y=A^{-1}y$
  
  となり、ご利益はない。

## 条件の分析

+ $y=ax+b$についての条件は

   $y_i=ax_i+b$ ($i=0,1,2$)

+ 行列で表すと、

    $Ak=y$ 
    
 ただし、$A=\left[\begin{matrix}x_0&1\\x_1&1\\x_2&1\end{matrix}\right]$,$k:=\left[\begin{matrix}a\\b\end{matrix}\right]$, $y:=\left[\begin{matrix}y_0\\y_1\\y_2\end{matrix}\right]$

+ 拡大係数行列 $[A|y]$の階数が$A$の階数$2$と同じでなければ解はない。

+ $rank A<3$は行ベクトルが一次従属ということなので

   $c_0[x_0,1,y_0]+c_1[x_1,1,y_1]+c_2[x_2,1,y_2]=0$
   
   第２成分をとれば $c_2=-c_1-c_2$よって
   
   $c_0[x_0-x_2,y_0-y_2]+c_1[x_1-x_2,y_1-y_2]=0$
   
   これは３点が一直線上にあることを意味する。
   
   したがって、３点が一直線上になければ $Ak=y$は解けない。


## 問題. 解けない方程式$Ak=y$の最良近似解を求めよ.
+ 「近似」をノルムで計測するとき問題は次のようになる：
+ 問題：$||Ak-y||$が最小となる$k$を求めよ.

## 復習1：直交射影

+ 復習. ${\mathbf R}^n$の部分空間$W$のベクトル ${\mathbf w}$と、$\mathbf y\in {\mathbf R}^n$ との距離$||\mathbf w-\mathbf y ||$の最小値は、
${\mathbf w}$が${\mathbf y}$の$W$への直交射影の時、最小となる.

+ 証明. 直交分解${\mathbb R}^n=W\oplus W^{\perp}$により${\mathbf y}={\mathbf y}'+{\mathbf y}''$ (${\mathbf y}'\in W$) と分解すると、
  
  $||{\mathbf w}-{\mathbf y}||^2=||{\mathbf w}-{\mathbf y}'||^2+ ||{\mathbf y}''||^2$
  
  よって、$\mathbf w=\mathbf y'$の時、最小値$||\mathbf y''||^2$ をとる。ここで ${\mathbf y}'$は${\mathbf y}$の$W$への直交射影に他ならない。

## 復習2：列ベクトル空間への直交射影

+ $W$が$(n,N)$行列$A$の列ベクトル空間の時、${\mathbf y}$の$W$への直交射影${\mathbf y}'$は、$A{\mathbf k}$と書ける。ただし、${\mathbf k}$は次の方程式の解：

    ${}^tA A{\mathbf k}={}^tA {\mathbf y}$


+ 証明. $W$の要素は$A{\mathbf k}$と表せる. 

　 $A{\mathbf k}$が${\mathbf y}$の直交射影となる条件は
     $A{\mathbf k}-{\mathbf y}$が$W$と直交すること、
   
   　 すなわち、任意の$\mathbf p\in \mathbb R^n$について$(A{\mathbf p},A{\mathbf k}-{\mathbf y})=(\mathbf p, {}^tAA{\mathbf k}-{}^tA{\mathbf y})=0$ となること. これは、
          ${}^tAA{\mathbf k}={}^tA{\mathbf y}$と同値.

## 系.  <font color="red">$||A{\mathbf k}-{\mathbf y}||$ は ${\mathbf k}=({}^tA A)^{-1}({}^tA {\mathbf y})$で最小となる.</font>

+ 前回の素数nに対する 1/n の周期のグラフで、周期>0.7 n の部分の方程式を最小二乗法で求めよ

## データ（第２回より）

```python
def cycle(n,x,k):
    history=[x]
    y=x
    y=(k*y)%n
    while not (y in history):
        history.append(y)
        y=(k*y)%n
    if y==0:
        return [0]
    else:
        u=y
        orbit=[u]
        y=(k*y)%n
        while y!=u:
            orbit.append(y)
            y=(k*y)%n
        return orbit

def shuuki(n,k):
    return len(cycle(n,1,k)) 

N=1000
import numpy as np
dfD0=pd.DataFrame([[n,shuuki(n,10)] for n in range(1,N+1) ],columns=["x","y"])

from sympy import isprime
condPrime=np.array([isprime(n) for n in dfD0.x])

dfD=dfD0[np.all([dfD0.y>0.7*dfD0.x,condPrime],axis=0)]

px.scatter(dfD,x="x",y="y")
```
